# Training and Testing Set
Before we move any forward with our data, we should at fist separate our dataset to the train and test set. 
It is necessary because the if we do not split the data set into train and test set, instead train the model with entire data then our model may overfit the data and we donot know whether it generalize to the other test cases or not. 
So by splitting the data set into train and test set, we train the model only with the train data set and later we test how well our model can predict the price of housing by using the test set. By doing so, we can actually see if the model can generalize predict values of unseen housing data or not. 

```python
# Let's first define the features and target variable
X = df.drop('median_house_value', axis=1) # everything except the house value
y = df['median_house_value'] # only the house value which is our target value

# splitting
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 
train_set, test_set = train_test_split(df, test_size=0.2) 
# Though the above statement does split our data to train and test. 
# It might be skewed and not represent the same proportionality of the data set
```

```python
test_set.describe()
```

```python
# Let's say we want to split data on basis of income. So that our test and train data contains the data of 
# households of all the income level

df["income_cat"] = pd.cut(df["median_income"], bins=[0., 1.5, 3.0, 4.5, 6., np.inf], labels=[1, 2, 3, 4, 5])

# These categories are represented as 
df["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.xlabel("Income category")
plt.ylabel("Number of districts")
plt.show() 
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
train_set, test_set = train_test_split(df, test_size=0.2, stratify=df['income_cat'], random_state=42)
```

```python
test_set.describe()
```

```python
# we are not gonna use the income_cat from nowonwards. so let's drop this 
for set in (train_set, test_set):
    set.drop('income_cat', inplace=True, axis = 1)
```

# Handling text and categorical data
The ocean_proximity contains the categorical data. So we need to convert that to some sort of numerical form. 
I think, onehot encoding will be great. 

**One-hot Encoding:**: It generates one binary attribute per category: one attribute equal to 1 when the category is "<1H OCEAN" (and 0 otherwise), another attribute equal to 1 when the category is "INLAND" (and 0 otherwise), and so on. This is called one-hot encoding because only one attribute will be equal to 1 (hot), while the others will be 0 (cold).

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df = train_set.copy()
# Create an instance of OneHotEncoder with sparse_output=False
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform the "ocean proximity" column
ocean_proximity_encoded = encoder.fit_transform(df[['ocean_proximity']])

# # Convert the result into a DataFrame
ocean_proximity_encoded_df = pd.DataFrame(ocean_proximity_encoded, columns=encoder.get_feature_names_out(['ocean_proximity']))
# ocean_proximity_encoded_df

# # Drop the original "ocean_proximity" column
df.drop(['ocean_proximity'], axis=1, inplace=True)

# Reset the index for both DataFrames
df.reset_index(drop=True, inplace=True)
ocean_proximity_encoded_df.reset_index(drop=True, inplace=True)

# # Concatenate the one-hot encoded DataFrame with the original DataFrame
df = pd.concat([df, ocean_proximity_encoded_df], axis=1)
df
```
